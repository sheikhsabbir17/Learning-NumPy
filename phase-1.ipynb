Array Reshaping


```python
import numpy as np

new_arr = np.arange(12)
print("Original Array: \n", new_arr)

reshaped = new_arr.reshape((3,4))
print("Reshaped Array: \n", reshaped)

flattened = reshaped.flatten()
print("Flattened Array: \n", flattened)

raveled = reshaped.ravel()
print("Ravelled Array: \n", raveled)
```
